```python
#Import Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib

import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8)

 #Adjusts the configuration of the plots


#Read in the data
import gdown
url= 'https://drive.google.com/uc?id=1pyFwBrvwPVVKZjXcRwxwwrPIxyXiEZjt'
output='downloaded_file.csv'
gdown.download(url,output, quiet=False)

df= pd.read_csv('https://drive.google.com/uc?id=1pyFwBrvwPVVKZjXcRwxwwrPIxyXiEZjt')
```

```python
#Data Exploration
df.head()
```

```python
#Data Cleaning
df.isnull().sum()
df.fillna('')
```

```python
#Adjusting Data types
df['budget'].fillna(0, inplace = True)
df['gross'].fillna(0, inplace = True)

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
df.sort_values(by=['gross'], inplace= False, ascending = False)
```

```python
#Hypothesis, Budget will have a high correlation to Gross

#Budget vs Gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Budget in Millions')
plt.ylabel('Gross in Millions')
```

```python
#RegPlot for Budget vs Gross
sns.regplot(x='budget' , y= 'gross' , data=df, scatter_kws = {'color' : 'red'}, line_kws = {'color': 'black'})
```

```python
#Correlation
df.corr(method='pearson')
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot= True)
plt.title('Correlation Matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
```

```python
df.sort_values(by=['country'], inplace=True, ascending= False)
```

```python
#Hypothesis 2: Some companies tend to produce higher revenue than others
import pandas as pd
from scipy.stats import f_oneway

# Assuming your DataFrame is named 'df'
# Select relevant columns: 'company' and 'gross'
data_to_analyze = df[['company', 'gross']].dropna()

# Create a dictionary to store the data for each company
company_data = {}
for company, group_data in data_to_analyze.groupby('company')['gross']:
    company_data[company] = group_data.values

# Perform one-way ANOVA
f_statistic, p_value = f_oneway(*company_data.values())

# Display the results
print(f'F-statistic: {f_statistic}\nP-value: {p_value}')
```

```python
#Confidently reject the null hypothesis, there is a significant difference in gross revenue among different companie.
```
